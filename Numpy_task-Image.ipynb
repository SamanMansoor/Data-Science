# **Task Numpy Image**

```python
import cv2

import numpy as np

from PIL import Image

im_path = "/content/baboon.png"



im_array = cv2.imread(im_path)
display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))

# Define the range of red color in BGR format
lower_red = (0, 0, 50)
upper_red = (50, 100, 255) 
# Threshold the image to get a binary mask of red pixel
mask = cv2.inRange(im_array, lower_red, upper_red)
# Replace all red pixels with black pixels
im_array[mask != 0] = (0, 0, 0)
display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))
```
