```python
# Define the column names as strings
columns = ['Loan_ID', 'Gender', 'Married', 'Dependents', 'Graduate', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Loan_Status']

# Define a function to remove outliers based on the IQR method
def remove_outliers(df_from_excel, columns, threshold=1.5):
    clean_df = df_from_excel.copy()
    for column in columns:
        Q1 = df_from_excel[column].quantile(0.25)
        Q3 = df_from_excel[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR
        clean_df = clean_df[(clean_df[column] >= lower_bound) & (clean_df[column] <= upper_bound)]
    return clean_df

# Specify the numeric columns you want to check for outliers
numeric_columns = df_from_excel.select_dtypes(include=[int, float]).columns

# Remove outliers from the entire dataset
cleaned_df = remove_outliers(df_from_excel, numeric_columns)

# Check the shape of the new DataFrame to see how many rows were removed
print(f"Original DataFrame shape: {df_from_excel.shape}")
print(f"DataFrame shape after removing outliers: {cleaned_df.shape}")
```

# **Exploratory Data Analysis (EDA)**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the percentage of female applicants with approved loans
total_female_applicants = (df_from_excel['Gender'] == 2).sum()
approved_female_loans = ((df_from_excel['Gender'] == 2) & (df_from_excel['Loan_Status'] == 'Y')).sum()
percentage_approved_female_loans = (approved_female_loans / total_female_applicants) * 100

# Create a Seaborn scatter plot to display the result
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sns.set_color_codes("pastel")
sns.scatterplot(x=['Female'], y=[percentage_approved_female_loans], color='b', s=100)  # Adjust 's' for marker size
plt.xlabel('Gender')
plt.ylabel('Percentage Approved')
plt.title('Percentage of Female Applicants with Approved Loans')
plt.show()

# Print the percentage of female applicants with approved loans
print(f"Percentage of female applicants with approved loans: {percentage_approved_female_loans:.2f}%")
```

```python
# Calculate the average income of all applicants
average_income = df_from_excel['ApplicantIncome'].mean()

# Create a curve line plot to visualize the average income
x = np.linspace(0, 1, 100)  # Generating x values from 0 to 1
y = [average_income] * len(x)  # Generating y values as the average income

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Average Income', color='b', linestyle='-', marker='o')
plt.xlabel('Category')
plt.ylabel('Income')
plt.title('Average Income of All Applicants')
plt.legend()
plt.grid()
plt.show()

# Print the average income in percentage format
print(f"Average income of all applicants: {average_income:.2%}")
```

```python
# Calculate the average income of all graduate applicants (where 'Graduate' is 'Yes=1')
average_income_graduate = df_from_excel[df_from_excel['Graduate'] == 1]['ApplicantIncome'].mean()
# Print the average income

print(f"Average income of graduate applicants: ${average_income_graduate:.2f}")

# Create a histogram to visualize the income distribution of graduate applicants
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sns.histplot(data=df_from_excel[df_from_excel['Graduate'] == 1], x='ApplicantIncome', color='b', bins=20)  # Adjust the number of bins as needed
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Income Distribution of Graduate Applicants - Histogram')
plt.show()
```

```python
# Filter the data for graduate applicants
graduate_data = df_from_excel[df_from_excel['Graduate'] == 1]

# Calculate the total number of graduate applicants
total_graduate_applicants = len(graduate_data)

# Calculate the number of graduate applicants with an approved loan status
approved_graduate_applicants = len(graduate_data[graduate_data['Loan_Status'] == 'Y'])

# Calculate the percentage of graduate applicants with approved loans
percentage_approved_graduate_loans = (approved_graduate_applicants / total_graduate_applicants) * 100

# Print the percentage
print(f"Percentage of graduate applicants with approved loans: {percentage_approved_graduate_loans:.2f}%")

# Create a pie chart to visualize the percentage of graduate applicants with approved loans
labels = ['Approved', 'Not Approved']
sizes = [percentage_approved_graduate_loans, 100 - percentage_approved_graduate_loans]
colors = ['green', 'red']
explode = (0.1, 0)  # Explode the 'Approved' slice for emphasis

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Graduate Applicants with Approved Loans')
plt.show()
```
