# **Cleaning and preprocessing the data**

```python
df.isnull().any()
```

```python
# find the null values
df.isnull().sum()
```

```python
df.dropna()
```

```python
Dup_Rows = df[df.duplicated()]

print("\n\nDuplicate Rows : \n {}".format(Dup_Rows))

```

```python
df.drop_duplicates()
print('\n\nResult DataFrame after duplicate removal :\n', df.head(n=5))
```

```python
#Remove Negaitive Value
df[(df['Quantity'] >= 0) | (df['Quantity'].isnull())]

```

**Merging and aggregating the data to create summary tables and metrics, such as total sales by product or customer.**

```python
df['year'] = pd.to_datetime(df['InvoiceDate']).dt.year
df.head()
```

```python
df['Sales'] = df['Quantity'].astype('int') * df['Price'].astype('float')
```

```python
df.groupby(['year']).sum()
```

# **Analyzing the data to identify trends and patterns**

```python
import matplotlib.pyplot as plt

year = range(0,1)
print(year)

plt.bar(year,df.groupby(['year']).sum()['Sales'])
plt.xticks(year)
plt.ylabel('Total Sales')
plt.xlabel('Per Year')
plt.show()
```

```python
df.groupby(['Country']).sum()
```

```python
import matplotlib.pyplot as plt

keys = [country for country, df in df.groupby(['Country'])]

plt.bar(keys,df.groupby(['Country']).sum()['Sales'])
plt.ylabel('Sales')
plt.xlabel('As per Year')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
#product sold the most
product = df.groupby('Customer ID')
quantity_ordered = product.sum()['Quantity']

keys = [pair for pair, df in product]
plt.scatter(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```
